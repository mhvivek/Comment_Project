```python
#Hi Vivek!! I fixed the space issue and read over the comments, but for some reason when I try to write them to a new file it says the file doesn't exist (which likely means I'm doing something wrong). I will try again tonight (probably around 8) but I just wanted to let you know where we are at.

#Hi! I got the fil writing to work (I think you were missing the "w" when you want to write to a file) and I also put in some extra sentences. Grace's comment is still quite short, so I will try to figure out where I can put in more sentences. Otherwise, I think that all we need to do is separate the comment out into paragraphs and comment our code

def final_pct_calculator(test_scores, pset_scores):
    '''calculates test + problem set combined score'''
    test = sum(test_scores)/len(test_scores)
    pset = sum(pset_scores)/len(pset_scores)
    avg = .75*pset + .25*test #weighting of scores
    return avg


def grade(pct):
    '''returns a grade when given a percent'''
    if pct >= 92.5:
        grade = "A"
    elif pct >= 89.5:
        grade = "A-"
    elif pct >= 86.5:
        grade = "B+"
    elif pct >= 82.5:
        grade = "B"
    elif pct >= 79.5:
        grade = "B-"
    elif pct >= 76.5:
        grade = "C+"
    elif pct >= 72.5:
        grade = "C"
    elif pct >= 69.5:
        grade = "C-"
    elif pct >= 66.5:
        grade = "D+"
    elif pct >= 62.5:
        grade = "D"
    elif pct >= 59.5:
        grade = "D-"
    else:
        grade = "F"
    return grade


def attendance(att, name, pronouns):
    '''creates sentence based on their attendance'''
    sent = ""
    if att > 8: #special message if they miss too many classes
        sent += pronouns[0].title() + " has missed " + str(att) + " classes, and must take responsibility for catching up on missed work. I am available before school and during office hours most days."
    elif att > 4:
        sent += pronouns[0].title() + " should make sure to keep up with work missed."
    return sent


def score_analyzer(test_scores, pset_scores, name, pronouns):
    '''returns how they do on tests compared to problem sets. Do they do well with an outlier or not try hard enough?'''
    info = ""
    top = False
    struggle = False
    if sum(test_scores) > 450: #if they did well on tests
        info += pronouns[0].title() + " scores excellent marks on tests (" + str(sum(test_scores)/len(test_scores)) + " average)"
        top = True
        if sum(pset_scores) > 450: # if they did well on both
            info += " as well as problem sets " + "(" + str(sorted(pset_scores)[0]) + " average)"
        if sorted(test_scores)[1] - sorted(test_scores)[0] > 10: #if there is one that is different by 10 or more
            info += " with one exception" + " (" + str(sorted(test_scores)[0]) + "). This shows me that " + name.split()[0] + " has mastered most of the target concepts in this class. I would be happy to go over the one test that was more challenging"
        if sum(test_scores)/5 - sum(pset_scores)/5 > 3:
            info += ", but finds problem sets more difficult (" + str(sorted(test_scores)[0]) + ") I would recommend setting aside one day for rereading code and checking for edge cases in problem sets"
        info += "."

    else:
        if sum(pset_scores) > 450: #if they did well on problem sets
            info += name.split()[0] + " solves problem sets with ease (" + str(sum(pset_scores)/len(pset_scores)) + " average)"
            top = True
            if sorted(pset_scores)[1] - sorted(pset_scores)[0] > 10: # one exception
                info += " with one exception" + "(" + str(sorted(pset_scores)[0]) + "). This shows me that " + name.split()[0] + " has mastered most of the target concepts in this class. I would be happy to go over the one problem set that was more difficult after or before school"
            if sum(pset_scores)/5 - sum(test_scores)/5 > 3: # if tests were much harder
                info += ", but finds tests more difficult (" + str(sum(test_scores)/len(test_scores)) + " average). If " + name.split()[0] + " isn't already, I would suggest preparing for tests by writing down the answers to review packets by hand. That way, " + pronouns[0] + " can simulate a test situation and itentify which areas are most challenging for " + pronouns[2]
            info += "."

        else: #if they didn't do well on either
            struggle = True
            info = pronouns[0].title() + " has struggled in both problem sets and test settings (averages being " + str(sum(pset_scores)/len(pset_scores)) + " and " + str(sum(test_scores)/len(test_scores)) + " respectively). I would love to hear what strategies you are using. Have you tried reviewing your Assignments or writing out the answers to the review packets by hand?"

    if sorted(test_scores)[-1] - sorted(test_scores)[-2] > 10 or sorted(pset_scores)[-1] - sorted(pset_scores)[-2] > 10: #if one score is a lot higher than all the rest - compares the difference of the highest score from the next highest
        info += ' ' + name.split()[0] + " has shown the ability to code well, as showcased by one score much higher than all the rest. I believe that " + pronouns[0] + " can consistently maintain those levels buy putting more effort into " + pronouns[2] + " work and I would be open to sitting down and trying to find ways for " + name.split()[0] + " to be more motivated."

    return [info, top, struggle] # top and struggle are used in the 'trend' function


def trend(test_scores, pset_scores, name, pronouns, top, struggle):
    '''returns if there were any trends in the scores'''
    info = ""
    if top:
        start = "Additionally, "
    if struggle:
        start = "Despite this, "
    over_time = []
    for i in range(5):
        over_time.append(test_scores[i] + pset_scores[i]) #combines test and problem set scores

    improve = True
    worsen = True
    for j in range(1, 5): #for each set of scores
        if over_time[j] < over_time[j-1]:
            improve = False #if they consistently improved
        if over_time[j] > over_time[j-1]:
            worsen = False #if they consistently declined
    if improve: #then...
        info += start + name.split()[0] + " has shown remarkable improvements over the course of the semester as shown by consistently improving scores. I hope that " + pronouns[0] + " continues to diligently work to make " + pronouns[1] + " code more clear and concise. Test and problem set scores will come in turn."
    if worsen:
        info += "Declining test and problem set scores lead me to believe that " + name.split()[0] + "'s focus has lessened over the course of the semester. I would love to know what I can do help you get back to where you were at the start of the year."
    return info


def final_sent(final_grade, name, pronouns):
    '''creates final sentence showing implication of final grade'''
    if final_grade == 'A': #unique final sentence depending on the grade
        sent = "Computer Science is a new skill for many students, and " + pronouns[0] + " has demonstrated " + pronouns[2] + " ability to learn quickly and excel in the course. " + pronouns[0].title() + " earns an A for the semester, and I look forward to continuing to work with " + pronouns[1] + " throughout the rest of the year."
    elif final_grade == 'A-':
        sent = "Computer Science is a new skill for many students, and " + pronouns[0] + " has demonstrated " + pronouns[2] + " ability to learn quickly and excel in the course. " + pronouns[0].title() + " should slow down and take a good look at " + pronouns[2] + " work before turning it in. Pay attention to detail! " + pronouns[0].title() + " earns an A- for the semester, and I look forward to continuing to work with " + pronouns[1] + " throughout the rest of the year."
    elif final_grade == 'B+':
        sent = "Computer Science is a new skill for many students, and " + pronouns[0] + " has demonstrated " + pronouns[2] + " ability to learn quickly and excel in the course. " + pronouns[0].title() + " should slow down and take a good look at " + pronouns[2] + " work before turning it in. Pay attention to detail! " + pronouns[0].title() + " earns an B+ for the semester, and I look forward to continuing to work with " + pronouns[1] + " throughout the rest of the year."
    elif final_grade == 'B':
        sent = "Computer Science is a new skill for many students, and it's clear how hard " + pronouns[0] + " has been working to manage this challenge. " + pronouns[2].title() + " hard work has paid off." + pronouns[2].title() + " final grade for the semester is a B, and I look forward to continuing to work with " + pronouns[1] + " throughout the rest of the year."
    elif final_grade == 'B-':
        sent = "Computer Science is a new skill for many students, and there have been many challenging assignments throughout the semester. Nevertheless, " + pronouns[2] + " final grade is a B-, demonstrating " + pronouns[2] + " hard work and perseverance. I look forward to continuing to work with " + pronouns[1] + " throughout the rest of the year."
    else:
        sent = "Computer science has been a challenge for " + pronouns[1] + ". " + pronouns[0].title() + " earns a " + final_grade + " for the semester, and I look forward forward to seeing " + pronouns[2] + " progress during the rest of the year."
    return sent


def homework(hw, name, pronouns): #hw is a number of assignments missed
    '''returns whether a student turned problem sets/other homework in on time'''
    if hw == 0:
        sent = pronouns[0].title() + " has practiced good habits by being on top of schoolwork and has completed all the homework assignments this year."
    elif hw == 1:
        sent = pronouns[0].title() + " has 1 missing homework assignment. If " + pronouns[0] + " is having trouble completing this assignment, I encourage " + pronouns[1] + " to come meet with me during my office hours."
    else: #if they missed more than one
        sent = pronouns[0].title() + " has a few missing homework assignments. If " + pronouns[0] + " is having trouble completing these assignments, I encourage " + pronouns[1] + " to come meet with me during my office hours. The homework is designed to help " + pronouns[1] + " excel in the class, and it is important that " + pronouns[0] + " completes it."
    return sent


def classwork():
    """returns the first semester course description"""
    course = "This semester in Computer Science 1, we learned about looping, conditionals, functions, lists, sets, and dictionaries. We also completed a unit on simulations and plotting. Students were encouraged to use their classmates as resources and participated in a coding challenge. Each unit involved a Problem Set based on the concepts learned in the Assignment as well as a pencil-and-paper test. The projects this semester were: simulating the game of craps, and a partner project on a bank customer service simulation."
    return course


def groupwork(small, large, name, pronouns): #small and large are numbers corresponding to participation in the corresponding type of group
    """returns comment on student participation"""
    if small - large > 5: # work a lot better in small groups
        group = name + " works well in small groups but should push " + pronouns[1] + "self to participate in large class discussions. I see that " + name.split()[0] + " has good ideas and is engaged, but " + pronouns[0] + " shouldn't hold back from sharing them."
    elif large - small > 5: # work a lot better in large groups
        group = name + " frequently asks clarifying questions in large class discussions but should push " + pronouns[1] + "self to participate in smaller group work. This type of participation is equally as valuable."
    else:
        if large > 6 and small <= 6:
            group = name + " works well in large groups but should focus on participating in small group discussions. This type of participation is equally as valuable."
        elif large <= 6 and small <= 6: # not high number in either
            group = "So far, I have noticed that " + name + " has trouble working with other students.  Taking risks can be beneficial in many different ways, and " + pronouns[0] + " should try to speak up more. I would hope that our class is a comfortable space for all of us to talk."
        elif large > 6 and small > 6:
            group = name + " frequently contributes to both full class and small group/partner discussions and activities. " + name.split()[0] + "'s ideas enrich both environments, are clearly stated, and try to bring others into the discussions. You are doing a good job right now, but keep making sure to monitor your airtime."
        if large <= 6 and small > 6: # high number in both
            group = name + " collaborates well in small groups but seems hesitant to speak up in class discussions. Try taking some risks!"
    return group


def main():
    with open("comments.csv") as com:
        header = com.readline()
        for line in com:
            line = line.split(",")
            name = line[0]
            pronouns = line[1].split("/")
            test_scores = [int(num) for num in line[2:7]]
            pset_scores = [int(num) for num in line[7:12]]
            att = int(line[12])
            hw = int(line[13])
            small = int(line[15])
            big = int(line[14]) #lines above break csv file into usable values - this could also be done using a dictionary
            pct = final_pct_calculator(test_scores, pset_scores)
            final = grade(pct)
            with open(f"{name.split()[0]}_{name.split()[1]}.txt", "w") as com: #creates file for each student
                comment = '' # start the comment from scratch
                #comment += classwork() + ' ' + groupwork(small, big, name, pronouns) + ' '
                if att > 0: #only execute the attendance function if they missed a class
                    comment += attendance(att, name, pronouns) + ' '
                comment += homework(hw, name, pronouns) + ' ' #adds homework to the comment
                analysis = score_analyzer(test_scores, pset_scores, name, pronouns)
                comment += analysis[0] + ' ' + trend(test_scores, pset_scores, name, pronouns, analysis[1], analysis[2]) #adds score analysis and trends
                comment.replace(".", ". ") #cleanup
                comment.replace('.  ', '. ') #cleanup
                com.write(classwork() + "\n")
                com.write(groupwork(small, big, name, pronouns))
                com.write(comment + '\n')
                com.write(final_sent(final, name, pronouns))


if __name__ == "__main__":
    main()

with open("Grace_Green.csv") as csvfile:
    for line in csvfile.readlines():
        print(line)
    print()
    print()
with open("Peter_Gillespie.csv") as csvfile:
    for line in csvfile.readlines():
        print(line)
    print()
    print()
with open("Annie_Watson.csv") as csvfile:
    for line in csvfile.readlines():
        print(line)
    print()
    print()
with open("Andy_Hou.csv") as csvfile:
    for line in csvfile.readlines():
        print(line)
    print()
    print()
```
